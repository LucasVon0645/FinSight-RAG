## Import Chat LLM config

```python
from importlib import resources as impresources
import finsight_rag.config as config
from finsight_rag.utils import load_yaml

chat_config_path = (impresources.files(config) / "chat_llm_config.yaml")
chat_config = load_yaml(chat_config_path)
print("Available chat LLM configurations:")
print(list(chat_config.keys()))
```

```python
gen_model = chat_config["model_id"]
temperature = chat_config["temperature"]
max_new_tokens = chat_config["max_new_tokens"]

print(f"Using generation model: {gen_model}")
print(f"Temperature: {temperature}")
print(f"Max new tokens: {max_new_tokens}")
```

```python
from finsight_rag.llms.llm_service import build_gemini_chat_llm

gemini_chat_llm = build_gemini_chat_llm(
    model_id='gemini-2.5-flash',
    temperature=temperature,
    max_new_tokens=max_new_tokens,
)
```
